# Examples

A closer look at `%%exercise`: isolation between cells, how errors are
reported, and why importing an already-loaded library inside the cell
doesn't reach back into the notebook's own copy of it.

```python
import script_widget
```

## A cell's variables don't leak out

`%%exercise` runs in a namespace that's fresh on every invocation, so a
name a cell defines is gone once the cell finishes:


```python
%%script
x = 42
for i in range(40):
    print(i)
asdf
```

A later `%%exercise` cell can't see it either -- referencing `x` again
raises a `NameError`. `%%exercise` catches that error and shows it in
the box rather than crashing the notebook:


```python
%%script
x
```

The same is true of any exception the cell raises -- it's reported with
its full traceback, not propagated to the notebook:


```python
%%script
1 / 0
```

## Imports are isolated too

`import matplotlib.pyplot as plt` inside the cell doesn't reach back into
the notebook's own matplotlib session -- each `%%exercise` cell runs in a
brand-new subprocess with its own `sys.modules`, so the import is
genuinely fresh. Mutating it -- here, the default line width -- has no
effect outside the cell:

```python
%%script
import matplotlib.pyplot as plt
plt.rcParams['lines.linewidth'] = 5
plt.plot([1, 3, 2])
```

```python
import matplotlib.pyplot as plt
plt.rcParams['lines.linewidth']
```

## stdout and stderr

Both streams are captured and shown separately:


```python
%%script
import sys
print('to stdout')
print('to stderr', file=sys.stderr)
```
